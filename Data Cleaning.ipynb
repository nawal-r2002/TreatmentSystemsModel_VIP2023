```python
import pandas as pd

data = pd.read_csv('Initial Drug-Drug Interactions.csv')
```

```python
data = data.drop_duplicates()
data = data.drop(columns=data.columns[[0,7,8,9,13,15,17,18]])
data = data.drop(columns=data.columns[12:16])
data
```

```python
for x in data.index:
    if data.loc[x, "Country where Event occurred"] == "Not Specified":
        data.drop(x,inplace=True)

data
```

```python
for x in data.index:
    if data.loc[x, "Suspect Product Names"] == "-":
        data.drop(x,inplace=True)

for x in data.index:
    if data.loc[x, "Concomitant Product Names"] == "-":
        data.drop(x,inplace=True)
        
data
```

```python
for x in data.index:
    if data.loc[x, "Patient Age"] == "Not Specified":
        data.drop(x,inplace=True)

for x in data.index:
    if data.loc[x, "Patient Weight"] == "Not Specified":
        data.drop(x,inplace=True)
data
```

```python
data['Observed Drug-Drug Interaction'] = None
col = data.pop('Observed Drug-Drug Interaction')
col2 = data.pop("Suspect Product Active Ingredients")
col3 = data.pop("Concomitant Product Names")
col4 = data.pop("Suspect Product Names")

data.insert(0, col.name, col)
data.insert(1, col2.name, col2)
data.insert(2, col4.name, col4)
data.insert(3, col3.name, col3)

data["Observed Drug-Drug Interaction"][data["Suspect Product Active Ingredients"].str.contains("Pioglitazone Hydrochloride")] = "Metformin Hydrochloride\Pioglitazone Hydrochloride"
data["Observed Drug-Drug Interaction"][data["Suspect Product Active Ingredients"].str.contains("Insulin Degludec")] = "Insulin Degludec\Liraglutide"
data["Observed Drug-Drug Interaction"][data["Suspect Product Active Ingredients"].str.contains("Insulin Glargine")] = "Insulin Glargine\Lixisenatide"
data["Observed Drug-Drug Interaction"][data["Suspect Product Active Ingredients"].str.contains("Glyburide")] = "Glyburide\Metformin"
```

```python
data = data.reset_index(drop=True)
data.to_csv("Interactions Database.csv", index=False)
```
